```python
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv(r'C:\Users\UPENDRA\Desktop\HAR_datset\csv_files\train.csv')
test = pd.read_csv(r'C:\Users\UPENDRA\Desktop\HAR_datset\csv_files\test.csv')
print(train.shape, test.shape)
```

```python
train.head()
```

```python
X_train = train.drop(['subject', 'Activity', 'ActivityName'], axis=1)
y_train = train.ActivityName
```

```python
X_test = test.drop(['subject', 'Activity', 'ActivityName'], axis=1)
y_test = test.ActivityName
```

```python
print('X_train and y_train : ({},{})'.format(X_train.shape, y_train.shape))
print('X_test  and y_test  : ({},{})'.format(X_test.shape, y_test.shape))
```

```python
import itertools
import matplotlib.pyplot as plt
from sklearn import metrics

plt.rcParams["font.family"] = 'DejaVu Sans'
def plot_confusion_matrix(cm,normalize=False,title='Confusion matrix',cmap=plt.cm.Blues):
    
    labels=['LAYING', 'SITTING','STANDING','WALKING','WALKING_DOWNSTAIRS','WALKING_UPSTAIRS']
    classes=labels
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def perform_model(y_test,y_pred):

    # calculate overall accuracy of the model
    accuracy = metrics.accuracy_score(y_true=y_test, y_pred=y_pred)
    print('Accuracy    : ',accuracy)
    
    # confusion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8,8))
    plt.grid(b=False)
    plot_confusion_matrix(cm , normalize=True, title='Normalized confusion matrix', cmap = plt.cm.Greens)
    plt.show()
    
    # get classification report
    print(' Classifiction Report :')
    classification_report = metrics.classification_report(y_test, y_pred)
    print(classification_report)
```

# 1. Logistic Regression with Grid Search

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
# start Grid search
parameters = {'C':[10**-3, 10**-2, 10**0, 10**2,10**3,10**4], 'penalty':['l2','l1']}
log_reg = LogisticRegression()
log_reg_grid = GridSearchCV(log_reg, param_grid=parameters, cv=5, verbose=1, n_jobs=-1)
grid_result=log_reg_grid.fit(X_train, y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
lr = LogisticRegression(C=1)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
perform_model(y_test,y_pred)
```

#  2. Linear SVC with GridSearch

```python
from sklearn.svm import LinearSVC
```

```python
parameters = {'C':[10**-3, 10**-2, 10**0, 10**2,10**3,10**4]}
lr_svc = LinearSVC()
lr_svc_grid = GridSearchCV(lr_svc, param_grid=parameters, n_jobs=-1, verbose=1)
grid_result = lr_svc_grid.fit(X_train, y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
lr_svc = LinearSVC(C=1)
lr_svc.fit(X_train,y_train)
y_pred = lr_svc.predict(X_test)
perform_model(y_test,y_pred)
```

# 3. Extra Tress  Classifier with GridSearch

```python
from sklearn.ensemble import ExtraTreesClassifier
params = {'n_estimators':[10,20,50,100,300,500],'max_depth':[5,15,30,50,100]}
etc = ExtraTreesClassifier()
etc_grid = GridSearchCV(etc, param_grid=params, n_jobs=-1)
grid_result = etc_grid.fit(X_train, y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
etc = ExtraTreesClassifier(max_depth=50,n_estimators=500)
etc.fit(X_train,y_train)
y_pred = etc.predict(X_test)
perform_model(y_test,y_pred)
```

# 4. xgbooost classifier with GridSearch

```python
import xgboost as xgb
param_grid= {'n_estimators':[10,20,50,100,300,500],'max_depth':[3,5,7,11,17,21]}
gbdt = xgb.XGBClassifier(objective='binary:logistic')
gbdt_grid = GridSearchCV(gbdt, param_grid=param_grid, n_jobs=-1)
grid_result = gbdt_grid.fit(X_train, y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
gbdt = xgb.XGBClassifier()
gbdt.fit(X_train,y_train)
y_pred = gbdt.predict(X_test)
perform_model(y_test,y_pred)
```
